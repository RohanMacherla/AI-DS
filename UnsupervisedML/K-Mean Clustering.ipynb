<h1>K-Mean Clustering</h1>



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Define the data points (A, B, C, D, E, F)
data = np.array([
    [2, 3],  # Point A
    [3, 3],  # Point B
    [6, 8],  # Point C
    [8, 8],  # Point D
    [7, 5],  # Point E
    [22,23]  # Point F
])
```

```python
# Perform K-means clustering with K=2
kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
 
# Get the cluster centers (centroids) and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Print the final centroids and cluster assignments
print("Final Centroids (Cluster Centers):")
print(centroids)
 
print("\nCluster Assignments:")
for i, label in enumerate(labels):
    print(f"Point {chr(65+i)}: Cluster {label+1}")
```

```python
# Plot the data points and the centroids
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o', label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100, label='Centroids')
# Annotate points
for i, point in enumerate(data):
    plt.text(point[0]+0.1, point[1]+0.1, f'{chr(65+i)}', fontsize=12)
 
plt.title('K-means Clustering Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('C:/AI-DS/UnsupervisedML/kmeans_dataset_with_outliers.csv')
df
```

```python
df = df.drop(columns='Label',axis=1)
df
```

```python
# Perform K-means clustering with K=2
kmeans = KMeans(n_clusters=2, random_state=0).fit(df)
 
# Get the cluster centers (centroids) and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Print the final centroids and cluster assignments
print("Final Centroids (Cluster Centers):")
print(centroids)
 
print("\nCluster Assignments:")
for i, label in enumerate(labels):
    print(f"Point {chr(65+i)}: Cluster {label+1}")
```

```python
# Plot the data points and the centroids
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100)
# Annotate points
for i, point in enumerate(df.values):
    plt.text(point[0]+0.1, point[1]+0.1, f'{chr(65+i)}', fontsize=12)

plt.title('K-means Clustering Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['Data Points', 'Centroids'])
plt.grid(True)
plt.show()
```

```python
X=df
```

```python
# List to store the within-cluster sum of squares (wcss) for different k values
wcss = []

# Test different values of k (number of clusters)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # inertia_ is the wcss

# Plot the elbow method graph
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('The Elbow Method')
plt.show()
```
